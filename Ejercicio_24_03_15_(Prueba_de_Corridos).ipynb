```python
import math
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x=[]
x1=[0.48, 0.591, 0.478, 0.166, 0.932]
print("  ", x1)
x.extend(x1)
x1=[0.926, 0.007, 0.551, 0.362, 0.533]
x.extend(x1)
print("  ", x1)
x1=[0.368, 0.437, 0.178, 0.411, 0.545]
print("X=",x1)
x.extend(x1)
x1=[0.787, 0.594, 0.086, 0.951, 0.772]
x.extend(x1)
print("  ", x1)
x1=[0.298, 0.199, 0.157, 0.401, 0.631]
x.extend(x1)
print("  ", x1)

gl=0.95
N=[]
n=x[0]
h=len(x)
for i in x:
  if i!=n:
    if n<i:
      N.append(1)
    if n>i:
      N.append(0)
  n=i

co=0
k=9
for i in N:
  if k==i:
    co+=0
  if k!=i:
     co+=1
  k=i

print("El numero total de natos n: ", h)

print("\n\tHo: Los numeros de la muestra Xi son independientes")
print("\tHa: Los numeros de la muestra Xi NO son independientes\n")
print("Conjunto S: ", N)
print("El numero de corridas es de", co)

Mc=(2*h-1)/3
var=(16*h-29)/90
Zo=(co-Mc)/var

za2=1.96
liminf=-za2
limsup=za2

print("\nEL intervalo de aceptaci贸n es de: [", liminf, ",", limsup, "]")
print(" ")
print("NUestro estadistico es:", Zo)
print("por lo tanto...\n")
if liminf<Zo and Zo<limsup:
  print("Se acepta Ho: Los numeros de la muestra son independientes")
else:
  print("Se rechaza Ho: ")

```
