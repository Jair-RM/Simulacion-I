```python
import math
import random as rnd
import numpy as np
import matplotlib.pyplot as plt

λ=15
n=10000
L=-1/λ
num=100
numeros=[]
while num!=0:
  u=rnd.random()
  x=L*np.log(u)
  numeros.append(x)
  num-=1

media=np.mean(numeros)
varianza= np.var(numeros)


print("La media es ", media)
print("1/λ es ", 1/λ)
print("La varianza es ", varianza)
print("1/λλ es ", 1/(λ*λ))
plt.hist(numeros)


```

```python
def cdf(numeros):
  n=len(numeros)
  x=np.sort(numeros)
  y=np.arange(0, len(numeros))/len(numeros)
  return x,y

X,Y=cdf(numeros)
plt.step(X,Y)
plt.step(X,Y, "cs", where="mid")
plt.xlim(0, 0.3)
plt.show()
```
