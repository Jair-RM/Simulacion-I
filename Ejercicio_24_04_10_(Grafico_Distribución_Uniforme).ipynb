```python
import math
import random as rnd
import numpy as np
import matplotlib.pyplot as plt

a=10
b=15
num=150

numeros=[]
while num!=0:
  u=rnd.random()
  x=u*(b-a)+a
  numeros.append(x)
  num-=1

media=np.mean(numeros)
varianza= np.var(numeros)


print("La media es ", media)
print("(a+b)/2 es ", (a+b)/2)
print("La varianza es ", varianza)
print("(b-a)2/12 es ", (pow(b-a,2))/12)
plt.hist(numeros)
```

```python
def cdf(numeros):
  n=len(numeros)
  x=np.sort(numeros)
  y=np.arange(0, len(numeros))/len(numeros)
  return x,y

X,Y=cdf(numeros)
plt.step(X,Y)
#plt.step(X,Y, "cs", where="mid")
plt.xlim(9, 16)
plt.show()
```

```python
import math
import random as rnd
import numpy as np
import matplotlib.pyplot as plt

x0=-2
gam=1.0
num=150

numeros=[]
while num!=0:
  u=rnd.random()
  x=gam*np.tan(np.pi*(u-0.5))+x0
  numeros.append(x)
  num-=1

media=np.mean(numeros)
varianza= np.var(numeros)


print("La media es ", media)

print("La varianza es ", varianza)

plt.hist(numeros)
```
