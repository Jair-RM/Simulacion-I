```python
import random as rnd
import numpy as np
import matplotlib.pyplot as plt

n=10
lam=2
L=-1/lam

def expo(lam=2, n=10):
   L=-1/lam
   list=[]
   for i in range (n):
      u=rnd.random()
      x=L*np.log(u)
      list.append(x)
   return list

g=expo(0.5, 1000)
print(g)
mu=np.mean(g)
sigma2=np.var(g)
#plt.hist(g, bins=50)
#bins es para hacer la gráfica más finita

plt.plot(g)
```
